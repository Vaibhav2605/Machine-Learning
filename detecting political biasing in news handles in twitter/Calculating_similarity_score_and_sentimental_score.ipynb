```python
import pandas as pd
import re
import operator
import numpy
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

# counting the frequency of hashtags 

```python
def Reverse(lst): 
    return [ele for ele in reversed(lst)] 

data=pd.read_csv("output_ndtv3.csv" ,usecols=["text","Mapped hashtags"]) 
Hashtags_count={}
for index, row in data.iterrows():
    li=row['Mapped hashtags'].replace('{',' ').replace(',',' ').replace('}',' ').replace('[',' ').replace(']',' ').split()
    list_set = set(li) 
    for val in list_set:
        if val not in Hashtags_count:
            Hashtags_count[val]=1
        else:
            Hashtags_count[val]+=1
        
```

```python
data_zee=pd.read_csv("output_zee3.csv" ,usecols=["text","Mapped hashtags"]) 
Hashtags_countzee={}
for index, row in data_zee.iterrows():
    li=row['Mapped hashtags'].replace('{',' ').replace(',',' ').replace('}',' ').replace('[',' ').replace(']',' ').split()
    list_set = set(li) 
    for val in list_set:
        if val not in Hashtags_countzee:
            Hashtags_countzee[val]=1
        else:
            Hashtags_countzee[val]+=1
```

```python
data_rep=pd.read_csv("output_republic3.csv" ,usecols=["text","Mapped hashtags"]) 
Hashtags_countrep={}
for index, row in data_rep.iterrows():
    li=row['Mapped hashtags'].replace('{',' ').replace(',',' ').replace('}',' ').replace('[',' ').replace(']',' ').split()
    list_set = set(li) 
    for val in list_set:
        if val not in Hashtags_countrep:
            Hashtags_countrep[val]=1
        else:
            Hashtags_countrep[val]+=1
```

#  find top 10 from each

```python
sorted_list = list(sorted(Hashtags_count.items(), key=operator.itemgetter(1)))
top_hash=sorted_list[-50:]
top_hash=Reverse(top_hash)
print(top_hash)
```

```python
sorted_listzee = list(sorted(Hashtags_countzee.items(), key=operator.itemgetter(1)))
top_hashzee=sorted_listzee[-50:]
top_hashzee=Reverse(top_hashzee)
print(top_hashzee)
```

```python
sorted_listrep = list(sorted(Hashtags_countrep.items(), key=operator.itemgetter(1)))
top_hashrep=sorted_listrep[-50:]
top_hashrep=Reverse(top_hashrep)
print(top_hashrep)
```

# Sentimental Score of Tweets 

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

## 1 NDTV

```python
analyser = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    #print(type(score))
    return score['compound']

dic_ndtv_score={}

for i in dic_ndtv.keys():
    total_score=0.0
    for j in dic_ndtv[i]:
        total_score+=sentiment_analyzer_scores(j)
    l=(len(dic_ndtv[i]))
    ans=total_score
    if l !=0 :
        ans=ans/len(dic_ndtv[i])
    if abs(ans*10)<1:
        ans=ans*10
    dic_ndtv_score[i]=ans
    
dic_ndtv_score
```

## 2 ZEE

```python
dic_zee_score={}

for i in dic_zee.keys():
    total_score=0.0
    for j in dic_zee[i]:
        total_score+=sentiment_analyzer_scores(j)
    l=(len(dic_zee[i]))
    ans=total_score
    if l !=0 :
        ans=ans/len(dic_zee[i])
    if abs(ans*10)<1:
        ans=ans*10
    dic_zee_score[i]=ans
dic_zee_score
```

## 3 Republic

```python
dic_rep_score={}

for i in dic_rep.keys():
    total_score=0.0
    for j in dic_rep[i]:
        total_score+=sentiment_analyzer_scores(j)
    l=(len(dic_rep[i]))
    ans=total_score
    if l !=0 :
        ans=ans/len(dic_rep[i])
    if abs(ans*10)<1:
        ans=ans*10
    dic_rep_score[i]=ans
dic_rep_score
```

```python
for i in list(dic_ndtv.keys()):
    cv = CountVectorizer()
    l =list(dic_ndtv[i])
    if len(l)==0:
        dict_ndtv_weighted[i]=list_zeros
        continue
    data_ndtv = cv.fit_transform(l)
    tfidf_transformer_ndtv = TfidfTransformer()
    
    # convert term-frequency matrix into tf-idf
    tfidf_matrix = tfidf_transformer_ndtv.fit_transform(data_ndtv)
    
    # create dictionary to find a tfidf word each word
    word2tfidf_ndtv = dict(zip(cv.get_feature_names(), tfidf_transformer_ndtv.idf_))
    
    dict_ndtv_weighted[i] = list(word2tfidf_ndtv.values())[0:10]
```
