#  PREPROCESSING ON THESE NEWS CHANNELS

```python
def exttweets(tweetss):         #to extract hashtags and create a list of them
    hasht=[]
    for i in range(len(tweetss)):
        h=re.findall(r"#(\w+)",tweetss[i])
        if(len(h) != 0):
            hasht.append(h)
    hashtags = [item for subl in hasht for item in subl]
    
    return hashtags
```

```python
def split_uc(s, keep_contiguous=False):
    
    string_length = len(s)
    is_lower_around = (lambda: s[i-1].islower() or 
                       string_length > (i + 1) and s[i + 1].islower())

    start = 0
    parts = []
    for i in range(1, string_length):
        if s[i].isupper() and (not keep_contiguous or is_lower_around()):
            parts.append(s[start: i])
            start = i
    parts.append(s[start:])

    return parts
```

```python
def tweprepro(lis):            #to preprocess on extracted hashtags list
    corpus = []
    for i in range(0, len(lis)):
        tex=  lis[i].lower()
        sw=set(stopwords.words('english')) #stemming  
        if(tex not in sw):
            corpus.append(tex)
    return corpus 
```

```python
def tweecount(corp):          #to create dictionary of extracted tweets with count
    hdict={}
    for i in range(len(corp)):
        c=0
        x=corp[i]
        for j in range(len(corp)):
            if(corp[j]==x):
                c=c+1
        hdict[x]=c 
    return hdict   
```
