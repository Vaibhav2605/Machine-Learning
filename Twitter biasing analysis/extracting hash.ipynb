# PREPROCESSING PER NEWS CHANNEL

```python
def exttweets(tweetss):         #to extract hashtags and create a list of them
    hasht=[]
    for i in range(len(tweetss)):
        h=re.findall(r"#(\w+)",tweetss[i])
        if(len(h) != 0):
            hasht.append(h)
    hashtags = [item for subl in hasht for item in subl]
    
    return hashtags
```

```python
def tweprepro(lis):            #to preprocess on extracted hashtags list
    corpus = []
    for i in range(0, len(lis)):
        tex=  lis[i].lower()
        ps = PorterStemmer()
        tex = ps.stem(tex)
        corpus.append(tex)
    return corpus    
```

```python
def tweecount(corp):          #to create dictionary of extracted tweets with count
    hdict={}
    for i in range(len(corp)):
        c=0
        x=corp[i]
        for j in range(len(corp)):
            if(corp[j]==x):
                c=c+1
        hdict[x]=c 
    return hdict   
```
