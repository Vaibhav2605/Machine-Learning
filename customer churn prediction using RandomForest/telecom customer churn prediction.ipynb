```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('Churn.csv')
```

```python
df=df.dropna()    #dropping rows with null values
```

# Data preprocessing and cleaning

```python
le=LabelEncoder()    #used for categorical data
ds = df.apply(le.fit_transform)
```

```python
ds.head(5)
```

### finding correlation matrix  

used corr so as to find the correlation b/w churn and the features so as to drop the features with very low correlation(less then 1% in this case)

```python
ds.corr(method='pearson')
```

```python
col_drop=['customerID','gender','PhoneService','TotalCharges']
ds=ds.drop(col_drop,axis=1)
```

```python
ds.head(5)
```

```python
dss=ds.values
dss.shape
```

```python
x=dss[:,:16]
y=dss[:,16]
```

```python
print(x.shape,y.shape)
```

```python
print(x)
split = int(0.56*x.shape[0])
```

### normalizing data 

```python
from sklearn import preprocessing
```

```python
x=preprocessing.normalize(x)
```

```python
print(x)
```

# machine learning model

## using RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(x[:split],y[:split])
```

## performance of the model

```python
rf.score(x[split:],y[split:])
```
