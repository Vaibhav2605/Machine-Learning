```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dfx=pd.read_csv("/home/vaibhav/Desktop/linear reg marks prediction/Linear_X_Train.csv")
dfy=pd.read_csv("/home/vaibhav/Desktop/linear reg marks prediction/Linear_Y_Train.csv")
print(dfx.shape,dfy.shape)
```

```python
#conversion to numpy
x=dfx.values
y=dfy.values
print(type(x))
print(y)
```

```python
#visualization of data
plt.scatter(x,y)
plt.show
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x
```

```python
def error(x,y,theta):
    m=x.shape[0]
    error=0
    
    for i in range(m):
        hx=hypothesis(x[i],theta)
        error += (hx-y[i])**2
        
    return error

def gradient(x,y,theta):
    grad=np.array([0,0])
    m=x.shape[0]
    
    for i in range(m):
        hx=hypothesis(x[i],theta)
        grad[1] += (hx-y[i])*x[i]
        grad[0] += (hx-y[i])
    return grad/m   
```

```python
def gradientDescent(x,y,lr= 0.005):
    theta=np.array([0.1,0.2])
    
    error_list=[]
    
    for i in range(1000):
        e=error(x,y,theta)
        error_list.append(e)
        grad = gradient(x,y,theta)
        theta[0]=theta[0]-lr*grad[0]
        theta[1]=theta[1]-lr*grad[1]
        
    return theta,error_list
```

```python
f_theta,error_list=gradientDescent(x,y)
```

```python
print(f_theta)
```

```python
plt.plot(error_list)
```
