# pre processing data

#### 1. handling the date of journey -  

taking only months from the date

```python
def months(X):
    m=X.strip()
    mm=m.split("/")
    mo=mm[1]
    return mo
    
```

```python
df_new['Date_of_Journey']=df_new['Date_of_Journey'].apply(months)
dft_new['Date_of_Journey']=dft_new['Date_of_Journey'].apply(months)
```

```python
df_new['Date_of_Journey']=df_new['Date_of_Journey'].astype(str)
dft_new['Date_of_Journey']=dft_new['Date_of_Journey'].astype(str)
```

#### 2. changing the time in dep_time in night / morning  

```python
def deptime(x):
    x=x.strip()
    m=int(x.split(":")[0])
    if(m>=12):
        x='Evening'
    else:
        x='Morning'
    return x    
```

```python
df_new['Dep_Time']=df_new['Dep_Time'].apply(deptime)
```

```python
dft_new['Dep_Time']=dft_new['Dep_Time'].apply(deptime)
```

```python
df_new.head()
```

```python
dft_new.head()
```

#### 3.Changing the duration to minutes  

```python
def timemin(x):
    x = x.strip()
    total=x.split(' ')
    to=total[0]
    hrs=(int)(to[:-1])*60
    if((len(total))==2):
          mint=(int)(total[1][:-1])
          hrs=hrs+mint
    x=str(hrs)
    return x
```

```python
df_new['Duration']=df_new['Duration'].apply(timemin)
```

```python
dft_new['Duration']=dft_new['Duration'].apply(timemin)
```

#### 4.total stops  

```python
def tstops(x):
    if(x=='non-stop'):
        x=str(0)
    else:
        x.strip()
        stps=x.split(' ')[0]
        x=stps
    return x

df_new['Total_Stops']=df_new['Total_Stops'].apply(tstops)
dft_new['Total_Stops']=dft_new['Total_Stops'].apply(tstops)
```
